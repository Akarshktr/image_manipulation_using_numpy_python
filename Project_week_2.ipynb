# Image Editor using Numpy

matplotlib -> plot the image

```python
from skimage import io, color
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage.util import invert
from PIL import Image
```

```python
class Image(): # image class
    def __init__(self,np_image=None):
        self.np_image = np_image
        
    def Load_Image(self):      #Method for load the image before doing all the operation
        self.np_image = io.imread('datascience.jpg')    # inputed image path should be given correctly
        print(self.np_image.shape)        # print the shape of the image
        print("Thank You ! for Load the Image Now You Can Perform many Operation: \n")
        fig, ax = plt.subplots()
        plt.imshow(self.np_image) # print the image 
        plt.show()
        
        
    def Edit_Image(self):
        while True:
            # user will enter the chice for for image editor
            print("Press 1 for Horizontal Flip : ")
            print("Press 2 for Vertical Flip : ")
            print("Press 3 for Invert Colours : ")
            print("Press 4 for Convert To Greyscale : ")
            print("Press 5 for Convert to B&W : ")
            print("Press 6 for Back to main Menu :")
            choice = int(input("Enter your choice:\n"))
            if choice == 1:
                np_image_horizontal_flip = self.np_image[::-1, :]    #horizontal flip
                fig, ax = plt.subplots()
                plt.imshow(np_image_horizontal_flip)
                plt.show()
            elif choice == 2:
                np_image_vertical_flip = self.np_image[:, ::-1]     # vertical flip
                fig, ax = plt.subplots()
                plt.imshow(np_image_vertical_flip)
                plt.show()
            elif choice == 3:
                np_image_inverted = self.np_image[:, :, ::-1]     #block  color inversion of an image 
                fig, ax = plt.subplots()
                plt.imshow(np_image_inverted)
                plt.show()
            elif choice == 4:
                np_image_greyscale = np.mean(self.np_image, axis=-1)
                fig, ax = plt.subplots()
                plt.imshow(np_image_greyscale, cmap='Greys')
                plt.show()
              
            elif choice == 5:
                np_image_greyscale1 = np.mean(self.np_image, axis=-1)
                np_image_blackandwhite=np_image_greyscale1.copy()
                np_image_blackandwhite[np_image_blackandwhite < 128] = 0
                np_image_blackandwhite[np_image_blackandwhite >= 128] = 255
                fig, ax = plt.subplots()
                plt.imshow(np_image_blackandwhite, cmap='Greys')
                plt.show()
            elif choice == 6:
                break  # it will break the loop
            else:
                print("Please Enter a valid Choice\n")  
                
    def Save_Image(self): # it will save the image in particular provided folder
        io.imsave(fname="C:\\Users\\Akarsh katiyar\\\\DSAT\\saved_image\\save.png", arr=self.np_image)
        print("IMAGE SAVED!!\n")
        
            
    
    
```
